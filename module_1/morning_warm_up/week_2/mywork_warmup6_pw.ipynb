```python
### Open the CO2 notebook we have worked on 
import pandas as pd
import numpy as np
table = pd.read_csv("co2_2.csv", header=None)
table.head()
```

```python
#### The goal is to use the month and year attributes/features as a categorical variable 
##to use in multiple regression
#- Do you understand why it's problematic to leave it as an integer type? 
## Can't interpret a unit increase in month when regressing as integer. 
#### 1. Load the co2 data 

#### 2. Follow all steps from previous notebook to get it to the state where we have 4 columns (year, month, day, CO2)
table = table.iloc[:,[0,1,3]]
```

```python
table['day']=15
table.columns = ['Year', 'Month', 'CO2', 'Day']
table.head()
```

```python
#### 3. Transform month and year to categorical variables 
table = pd.concat([table, pd.get_dummies(table['Year'])], axis=1)
table = pd.concat([table, pd.get_dummies(table['Month'])], axis=1)
table.head()
```

```python
#### 4. For practice purposes only let's create a function that gets a df and a column name and outputs 
## a normalized column - a column of our choice --> here we'll test the CO2 column 
## use the np.mean function and np.std `
table['CO2']= table['CO2'].replace(-99.99, np.nan)
table.head()
```

```python
table['Year'].values.mean()
```

```python
table1 = table.dropna(axis = 0, how ='any').copy()
table1.head()
```

```python
table1['co2avg']=table1['CO2'].mean()
table1.head()
```

```python
def normal(df,col):
    df['avg']=df[col].mean()
    df['s']=df[col].std()
    return df.avg, df.s
        
normal(table1,'CO2')
```
