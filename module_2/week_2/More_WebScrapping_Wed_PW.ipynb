```python
def process_page(page):
    tables = page.find_all('table')
    table =tables[0]
    rows = table.find_all('tr')
    datas = []
# # for i, row in enumerate(rows):
# #     if i != 0:
    for r in rows[1:]: # [1:] to exclude header
        data = []
        id = r.find('td').get_text()
        data.append(id)
        date = r.find('a').get_text()
        data.append(date)
        name = r.find('b').get_text()
        data.append(name)
        dat = r.find_all('td', class_="data")
        for d in dat[1:]:
            data.append(d.get_text())
        datas.append(list(data))
    return datas
```

```python
from os.path import join
movies_data = []
for i in range(1, MAX_PAGE+1, 100):
    resp = requests.get(join(BASE_URL, str(i)))
    soup = BeautifulSoup(resp.text,'html.parser')
    movies_data += process_page(soup)

len(movies_data)
```

```python
conn = sqlite3.connect(':memory:')
c = conn.cursor()
```
